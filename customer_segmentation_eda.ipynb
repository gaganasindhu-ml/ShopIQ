```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Mall_Customers.csv')
```

```python
print("Shape:", df.shape)
print("\nColumn names:", df.columns.tolist())
print("\nData types:")
print(df.dtypes)
print("\nFirst 5 rows:")
print(df.head())
print("\nBasic statistics:")
print(df.describe().round(1))
print("\nMissing values:")
print(df.isnull().sum())   # should all be 0
```

```python
plt.figure(figsize=(5, 4))
df['Gender'].value_counts().plot(kind='bar',
    color=['#534AB7','#1D9E75'], edgecolor='white')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

print("Gender counts:")
print(df['Gender'].value_counts())
```

```python
plt.figure(figsize=(6,4))
plt.hist(df['Age'], bins=15, color='#534AB7', edgecolor='white', alpha=0.8)
plt.title("Age Distrubuation of Customers")
plt.xlabel("Age")
plt.ylabel("count")
plt.tight_layout()
plt.show()

print("\nAge Statistics:")
print(f" Yougest: {df['Age'].min()}")
print(f" oldest: {df['Age'].max()}")
print(f" Average: {df['Age'].mean():.1f}")
```

```python
plt.figure(figsize=(6, 4))
plt.hist(df['Annual Income (k$)'], bins=15,
         color='#1D9E75', edgecolor='white', alpha=0.8)
plt.title('Annual Income Distribution')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Number of Customers')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.hist(df['Spending Score '], bins=15,
         color='#D85A30', edgecolor='white', alpha=0.8)
plt.title('Spending Score Distribution')
plt.xlabel('Spending Score')
plt.ylabel('Number of Customers')
plt.tight_layout()
plt.show()
```

```python
# ════════════════════════════════════════════════════
# EDA STEP 6 — THE KEY PLOT
# Income vs Spending — this is where the magic is
# ════════════════════════════════════════════════════
plt.figure(figsize=(8, 6))
colors_gender = df['Gender'].map({'Male':'#534AB7','Female':'#D85A30'})
plt.scatter(df['Annual Income (k$)'],
            df['Spending Score (1-100)'],
            c=colors_gender, alpha=0.7, s=60)

# legend manually
plt.scatter([], [], color='#534AB7', label='Male')
plt.scatter([], [], color='#D85A30', label='Female')
plt.legend()

plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Income vs Spending Score — colour by Gender')
plt.tight_layout()
plt.show()
```

```python

# ════════════════════════════════════════════════════
# EDA STEP 7 — Correlation
# Do higher income people spend more?
# ════════════════════════════════════════════════════
corr = df[['Age','Annual Income (k$)',
           'Spending Score ']].corr().round(2)
print("\nCorrelation matrix:")
print(corr)

plt.figure(figsize=(5, 4))
sns.heatmap(corr, annot=True, cmap='coolwarm',
            fmt='.2f', linewidths=0.5)
plt.title('Correlation between features')
plt.tight_layout()
plt.show()
```

```python
# ════════════════════════════════════════════════════
# EDA STEP 7 — Correlation
# Do higher income people spend more?
# ════════════════════════════════════════════════════
corr = df[['Age','Annual Income (k$)',
           'Spending Score ']].corr().round(2)
print("\nCorrelation matrix:")
print(corr)

plt.figure(figsize=(5, 4))
sns.heatmap(corr, annot=True, cmap='coolwarm',
            fmt='.2f', linewidths=0.5)
plt.title('Correlation between features')
plt.tight_layout()
plt.show()
```

```python
# ════════════════════════════════════════════════════
# EDA STEP 8 — Age vs Spending
# ════════════════════════════════════════════════════
plt.figure(figsize=(7, 5))
plt.scatter(df['Age'],
            df['Spending Score '],
            color='#EF9F27', alpha=0.7, s=60)
plt.xlabel('Age')
plt.ylabel('Spending Score')
plt.title('Age vs Spending Score')
plt.tight_layout()
plt.show()
```
